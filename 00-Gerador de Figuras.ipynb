```python
x = np.linspace(0.8, 2.7, 100)
def f(x):
    y = -1*(x-1.0)*(x-2.0)*(x-2.5) + 1.0
    return y
```

```python
x = np.linspace(0.8, 2.7, 100)
dx = np.linspace(0.8, 2.0, 100)
def f(x):
    y = -x**2 + 4 * x - 1.75 
    return y
def dfdx(x):
    y = 1.2 * x + 0.21
    return y
```

```python
y = f(x)
dy = dfdx(dx)
xp = 1.4
yp = f(xp)
print(yp)
```

```python
xmin, xmax, ymin, ymax = 0, 3, 0, int(y.max()+1)
print(xmin, xmax, ymin, ymax)
print(dx)
print(dy)
```

```python
fig = plt.figure()
ax = plt.axes()
fig.patch.set_facecolor('#ffffff')
ax.plot(x,y)
ax.plot(dx,dy)
ax.plot(xp, yp, 'ro')
ax.set(xlim=(xmin-1, xmax), ylim=(ymin-1, ymax), aspect='equal')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('$x$', size=14, labelpad=-24, x=1.02)
ax.set_ylabel('$y$', size=14, labelpad=-21, y=1.02, rotation=0)
#x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
#y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
#ax.set_xticks(x_ticks[x_ticks != 0])
#ax.set_yticks(y_ticks[y_ticks != 0])
ax.set_xticks([])
ax.set_yticks([])
plt.text(1.00, 0.82, r"$ y = f(x) $", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=14)
plt.text(0.90, 0.95, r"A linha tangente tem inclinação $f'(x_0)$", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=10)
plt.text(0.80, 0.75, r"$(x_0, f(x_0))$", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=14)
```

* Figura 1.4

```python
x = np.linspace(1, 5, 100)
def f(x):
    y = (x - 1)*(x - 3)*(x - 5)
    return y
x01 = (18 - 5 * np.sqrt(2))/6
x02 = (18 + 5 * np.sqrt(2))/6
print(x01, x02)
print(f(x01), f(x02))
def f_(x):
    y = (x - 1)*(x - 3)*(x - 5) + (2*x - 2) + 1 
    return y
print(x01, x02)
print(f_(x01), f_(x02))
xp = np.array([1, 5])
yp = f_(xp)
print(xp)
print(yp)
dx1 = np.linspace(1.2, 2.4, 100)
dx2 = np.linspace(3.5, 4.7, 100)
dy1 = np.ones_like(dx1) * f_(x01)
dy2 = np.ones_like(dx2) * f_(x02)
def f1(x):
    y = 2 * x - 1
    return y

def f2(x):
    y = 2 * x + (f_(x01) - 2 * x01)
    return y
dy1 = f1(x)
dy2 = f2(x)
```

```python
y = f_(x)
```

```python
xmin, xmax, ymin, ymax = 0, 6, int(y.min() - 5), int(y.max() + 1)
print(xmin, xmax, ymin, ymax)
```

```python
fig = plt.figure()
ax = plt.axes()
fig.patch.set_facecolor('#ffffff')
ax.plot(x,y)
ax.plot(x,dy1,'g')
ax.plot(x,dy2,'g')
ax.plot(xp, yp, 'ro')
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax), aspect='equal')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('$x$', size=14, labelpad=-24, x=1.02)
ax.set_ylabel('$y$', size=14, labelpad=-21, y=1.02, rotation=0)
#x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
#y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
#ax.set_xticks(x_ticks[x_ticks != 0])
#ax.set_yticks(y_ticks[y_ticks != 0])
ax.set_xticks([])
ax.set_yticks([])
plt.text(0.95, 0.55, r"$ y = f(x) $", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=14)
plt.text(0.55, 0.40, r"$\frac{f(b)-f(a)}{b - a}$", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=14)
plt.text(0.35, 0.80, r"$f'(c)$", ha='right', va='top',
    transform=ax.transAxes,
         horizontalalignment='center', fontsize=14)

#plt.text(0.90, 0.09, r"b", ha='right', va='top',
#    transform=ax.transAxes,
#         horizontalalignment='center', fontsize=10)
```
